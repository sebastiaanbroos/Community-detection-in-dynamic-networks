```python
# average data out
def avg_data(d, n):
  new_d = []
  new_f = []
  halt = 0
  if float(d[0][-2]) > 90:
    for i in range(len(d[0])):
      if halt == 1:
        break
      if float(d[0][i]) >= 90:
        halt = 1
      new_d.append(d[0][i])
      new_f.append(d[1][i])
    d = [new_d, new_f]
  correctness = []
  frac = []
  som1 = 0
  som2 = 0
  for i in range(0, len(d[0]) - n, n):
    for j in range(n):
      som1 += float(d[0][i + j])
      som2 += float(d[1][i + j])
    correctness.append(som1 / n)
    frac.append(som2 / n)
    som1 = 0
    som2 = 0
  correctness.append(d[0][-1])
  frac.append(d[1][-1])
  return np.array(correctness).astype(float), np.array(frac).astype(float)
```

```python
# define function for creating histogram data
def create_hist_data(degrees):
  histogram_dict = {'0-50':0, '50-100': 0, '100-150':0, '150-200':0, '200-250':0, '250-300':0, '300-350':0, '350-400':0 }
  for node in degrees:
    if node <= 50:
      histogram_dict['0-50'] +=1
    elif node <= 100:
      histogram_dict['50-100'] +=1
    elif node <= 150:
       histogram_dict['100-150'] +=1
    elif node <= 200:
       histogram_dict['150-200'] +=1
    elif node <= 250:
       histogram_dict['200-250'] +=1
    elif node <= 300:
       histogram_dict['250-300'] +=1
    elif node <= 350:
       histogram_dict['300-350'] +=1
    else:
      histogram_dict['350-400'] +=1
  return histogram_dict
```
